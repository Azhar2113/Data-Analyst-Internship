# ‚úÖ Level 1 - Task 1: Top Cuisines


üîç Objective:
- Identify the top 3 most common cuisines served by restaurants.

- Calculate the percentage of restaurants that serve each of these cuisines.

- Visualize the result using a bar chart.



üü¶ Step 1: Import Libraries

```python
import pandas as pd
from collections import Counter
import matplotlib.pyplot as plt
```

üü¶ Step 2: Load the Dataset


```python
df = pd.read_csv("DS.csv") # Make sure Dataset.csv is in your Jupyter folder
```

üü¶ Step 3: Clean & Prepare Cuisine Data

```python
# Drop missing values in the 'Cuisines' column
cuisine_series = df['Cuisines'].dropna()

# Split multiple cuisines and clean spaces
all_cuisines = [c.strip() for entry in cuisine_series for c in entry.split(',')]
```

üü¶ Step 4: Count and Get Top 3 Cuisines

```python
# Count all cuisines
cuisine_counts = Counter(all_cuisines)

# Get top 3 most common
top_3 = cuisine_counts.most_common(3)
print("Top 3 Cuisines:\n", top_3)
```

üü¶ Step 5: Calculate Percentage

```python
# Total restaurants (for percentage)
total_restaurants = len(df)

# Calculate percentage
top_3_with_percent = [(cuisine, count, round((count / total_restaurants) * 100, 2)) for cuisine, count in top_3]

# Convert to DataFrame
top_3_df = pd.DataFrame(top_3_with_percent, columns=["Cuisine", "Count", "Percentage"])
print("\nTop 3 Cuisines with Percentage:\n")
print(top_3_df)
```

üü¶ Step 6: Plot Bar Chart

```python
plt.figure(figsize=(6,4))
plt.bar(top_3_df["Cuisine"], top_3_df["Count"], color="skyblue")
plt.title("Top 3 Cuisines by Count")
plt.xlabel("Cuisine")
plt.ylabel("Number of Restaurants")
plt.grid(True, linestyle='--', alpha=0.5)
plt.show()
```
