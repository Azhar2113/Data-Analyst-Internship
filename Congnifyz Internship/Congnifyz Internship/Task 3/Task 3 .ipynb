# ‚úÖ Level 1 - Task 3: Price Range Distribution

üîç Objective:
1. Show how restaurants are distributed across different price ranges

2. Create a bar chart for visualization

3. Calculate the percentage of restaurants in each price range

üü¶ Step 1: Import Required Libraries


```python
import pandas as pd
import matplotlib.pyplot as plt
```

üü¶ Step 2: Load the Dataset

```python
df = pd.read_csv("DS.csv")
```

üü¶ Step 3: Count and Percentage of Each Price Range

```python
# Count the number of restaurants in each price range
price_counts = df['Price range'].value_counts().sort_index()

# Calculate percentage
price_percent = (price_counts / len(df) * 100).round(2)

# Combine into a DataFrame
price_df = pd.DataFrame({
    'Price Range': price_counts.index,
    'Count': price_counts.values,
    'Percentage': price_percent.values
})

print("üî∏ Price Range Distribution:\n")
print(price_df)
```

üü¶ Step 4: Plot the Bar Chart

```python
# Create a bar chart
plt.figure(figsize=(6,4))
plt.bar(price_df['Price Range'].astype(str), price_df['Count'], color='orange')
plt.title("Distribution of Price Ranges")
plt.xlabel("Price Range")
plt.ylabel("Number of Restaurants")
plt.grid(True, linestyle='--', alpha=0.5)
plt.show()
```
